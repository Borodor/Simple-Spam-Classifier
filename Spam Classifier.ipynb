```python
import pandas as pd
```

```python
data = open('Email spam.txt')
df = pd.DataFrame()
```

```python
target = []
msg = []
for i in data.readlines():
    i = i.lower()
    target.append(i.split()[0])
    msg.append(' '.join(i.split()[1:]))
```

```python
target[0] = 'ham'
```

```python
df['data'] = msg
df['target'] = target
```

```python
df.head()
```

```python
all_words = []
for i in msg:
    for j in i.split():
        all_words.append(j.lower())
```

```python
all_words[0]
```

```python
import nltk
```

```python
all_words = nltk.FreqDist(all_words)
all_words.most_common(10)
```

```python
len(all_words)
```

```python
word_features = []
top_3000 = all_words.most_common()[0:10000]
for i in top_3000:
    word_features.append(i[0])
```

```python
def find_features(document):
    words = document
    features = {}
    for w in word_features:
        features[w] = (w in words)
    return features
```
